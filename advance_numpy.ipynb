Advance Numpy

```python
import numpy as np
```

```python
# NumPy arrays are called homogeneous because all elements must have the same data type.
sales = np.array([100,120,130,150])
print(f"Sales      : {sales}")
print(f"Shape      : {sales.shape}")
print(f"Data Type  : {sales.dtype}")
print(f"Dimension  : {sales.ndim}")
```

```python
a = [100, 120, 130, 150]
b = [80, 90, 100, 110]
c =  [90, 95, 100, 105]
sales_matrix = np.array([a,b,c])
print(f"Sales Matrix : \n\n{sales_matrix}\n\n")
print(f"Shape        : {sales_matrix.shape}")
print(f"Data Type    : {sales_matrix.dtype}")
print(f"Dimension    : {sales_matrix.ndim}")
```

```python
# np.zeros() - np.zeros(shape) creates an array filled with zeros.
zeros_array = np.zeros(4) # Float - Data Type
print(f"Zeros Array : {zeros_array}")
print(f"Shape       : {zeros_array.shape}")
print(f"Data Type   : {zeros_array.dtype}")
```

```python
# Integer Data Type
zeros_array = np.zeros(4,dtype=int) 
print(f"Zeros Array : {zeros_array}")
print(f"Shape       : {zeros_array.shape}")
print(f"Data Type   : {zeros_array.dtype}")
```

```python
# np.ones() - np.ones(shape) creates an array filled with ones.
ones = np.ones((2,3))
print(f"Ones Array  : \n\n{ones}\n\n")
print(f"Shape       : {ones.shape}")
print(f"Data Type   : {ones.dtype}")
print(f"Dimension   : {ones.ndim}")
```

```python
# np.arange() - np.arange(start, stop, step) works like Python’s range() but returns a NumPy array.
q1 = np.arange(1, 5)
print(f"Quaters (1,5)     :  {q1}")
q2 = np.arange(10,20,2)
print(f"Quaters (10,20,2) :  {q2}")
q3 = np.arange(1,100,5)
print(f"Quaters (1,100,5) :  {q3}")
```

```python
# np.linspace() - np.linspace(start, stop, num) creates evenly spaced values between start and stop.
f1 = np.linspace(1,5,2)
print(f"Fractions 1 (1,5,2)   :  {f1}")
f2 = np.linspace(10,20,2)
print(f"Fractions 2 (10,20,2) :  {f2}")
f3 = np.linspace(1,100,5)
print(f"Fractions 3 (1,100,5) :  {f3}")
```

```python
one = np.array([10, 20, 30])
two = np.array([1, 2, 3])
print(f"Array Operations")
print(f"Addition        : {one+two}")
print(f"Subtraction     : {one-two}")
print(f"Multiplication  : {one*two}")
print(f"Division        : {one/two}")
print(f"Reminder        : {one%two}")
```

Statistical Functions

```python
print(f"Sales               : {sales}")
print(f"Mean                : {np.mean(sales)}")
print(f"Median              : {np.median(sales)}")
print(f"Standard Deviation  : {np.std(sales)}")
print(f"Minimum             : {np.min(sales)}")
print(f"Maximum             : {np.max(sales)}")
print(f"Sumation            : {np.sum(sales)}")
```

```python
# Axis in NumPy - In data science, calculations are often performed row-wise or column-wise.
print(f"Sales Matrix     : \n\n{sales_matrix}\n\n")
print(f"Column-Wise Sum  : {np.sum(sales_matrix, axis=0)}")
print(f"Row-Wise Sum     : {np.sum(sales_matrix, axis=1)}")
```

```python
# Reshape
numbers = np.array([1,2,3,4,5,6])
reshaped = numbers.reshape(2,3)
print(f"Original Numbers : {numbers}\n")
print(f"Reshaped Numbers : \n\n{reshaped}")
```

```python
# Random Numbers - Random data is important for simulations and machine learning.
random_data = np.random.randint(1, 100, 10)
print(f"Random Data          : {random_data}")
print(f"Random Decimal Data  : {np.random.rand(5)}")
```

```python
# Matrix Operations
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(A + B)
```

```python
# Unique Values
value = np.array([1,2,4,2,3,3,3])
print(f"Original Values       : {value}")
print(f"Unique Values         : {np.unique(value)}")
print(f"Sort Values           : {np.sort(value)}")
print(f"Conditional Selection : {np.where(value > 2)}")
```
