Data - Visualization Using Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
days = [1,2,3,4,5]
tempreature = [30,32,35,33,31]
plt.plot(days,tempreature)
plt.xlabel("Days")
plt.ylabel("Tempreature")
plt.title("Tempreature Analysis")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Simple Line Plot")
plt.grid()
plt.show()
```

```python
# Line Plot - Parameters
plt.plot(x,y,color='green',linestyle='-.',marker='o',linewidth=1.5,markerfacecolor='yellow',markeredgecolor='black',alpha=0.5,label='X and Y')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Plot X/Y")
plt.legend()
plt.grid(color='gray',linestyle='--',linewidth=0.5)
plt.show()
```

```python
plt.style.use('grayscale')
months = ['Jan','Feb','Mar','Apr','May']
sales = [100,120,150,130,170]
plt.figure(figsize=(10,5))
plt.plot(months,sales,color='blue',linestyle='--',linewidth=3,marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='black',alpha=0.8,label='Sales')
plt.title("Monthly Sales Analysis")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.grid()
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]
city1 = [30,32,35,33,31]
city2 = [28,29,31,30,29]
plt.figure(figsize=(10,5))
plt.plot(days,city1,label='City A',color='blue',linestyle='--',linewidth=3,marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='black',alpha=0.8)
plt.plot(days,city2,label='City B',color='red',linestyle='--',linewidth=3,marker='o',markersize=10,markerfacecolor='pink',markeredgecolor='black',alpha=0.8)
plt.grid(color='gray',linestyle='--',linewidth=0.5)
plt.xlabel("Days")
plt.ylabel("Cities Data")
plt.title("City Data / Days Analysis")
plt.legend()
plt.show()
```

```python
# Scatter Plot
hours = [1,2,3,4,5]
marks = [40,50,60,70,80]
plt.scatter(hours,marks,color='red',s=50,marker='D',alpha=0.5,edgecolors='black')
plt.show()
```

```python
# Color Maps
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
plt.scatter(x,y,c=z)
plt.colorbar()
plt.show()
```

```python
h = [1,2,3,4,5,6,7]
m = [45,50,55,65,70,80,90]
plt.figure(figsize=(8,5))
plt.scatter(h,m,color='blue',s=200,marker='o',alpha=0.7,edgecolors='black')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.grid()
plt.show()
```

```python
# Multiple Scatter Groups
x1 = [1,2,3]
y1 = [2,3,4]
x2 = [5,6,7]
y2 = [7,8,9]

plt.scatter(x1,y1,color='red',label='Group A')
plt.scatter(x2,y2,color='blue',label='Group B')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar Plot
products = ['Laptop','Mobile','Tablet']
sales = [100,150,80]
colors = ['red','green','orange']
plt.bar(products,sales,color=colors,width=0.5,edgecolor='black')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Electronics Sales")
plt.grid(color='gray',linestyle='-.',linewidth=0.5,axis='y')
plt.show()
```

```python
products = ['Laptop','Mobile','Tablet']
sales = [100,150,80]
colors = ['red','green','orange']
plt.barh(products,sales,color=colors,edgecolor='black')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Electronics Sales")
plt.show()
```

```python
# Grouped Bar Plot
sales_24 = [100,120,80]
sales_25 = [130,150,110]
x = np.arange(len(products))
width = 0.35
plt.bar(x-width/2,sales_24,width,label='2024')
plt.bar(x+width/2,sales_25,width,label='2025')
plt.xticks(x,products)
plt.title("Electonics Sales")
plt.legend()
plt.show()
```

```python
# Stacked Bar Plot
boys = [40,50,60]
girls = [30,40,50]
classes = ['A','B','C']
plt.bar(classes,boys,label='Boys')
plt.bar(classes,girls,bottom=boys,label='Girls')
plt.xlabel("Classes")
plt.ylabel("Students")
plt.title("APB School Class 10")
plt.legend()
plt.show()
```

```python
# Adding Values On Bar
bars = plt.bar(products,sales)
for bar in bars:
    y = bar.get_height()

    plt.text(bar.get_x() + bar.get_width()/2,y,y,ha='center')
plt.show()
```

```python
plt.figure(figsize=(10,5))
bars = plt.bar(products,sales,color=['red','blue','green','orange'],edgecolor='black',alpha=0.8)
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,y,y,ha='center')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Products Sales Dashboard")
# plt.grid(axis='y')
plt.show()
```

```python
# Histogram
marks = [45,50,52,60,61,65,70,72,75,80,85]
plt.hist(marks,bins=5,color='skyblue',edgecolor='black',alpha=0.7,density = True)
```

```python
marks = np.array([45, 50, 52, 60, 61, 65, 70, 72, 75, 80, 85])
mean = np.mean(marks)
std_value = np.std(marks)
x = np.linspace(mean - 3*std_value, mean + 3*std_value, 100)
y = stats.norm.pdf(x, mean, std_value)
plt.figure(figsize=(8, 5))
plt.hist(marks, bins=5, density=True, alpha=0.6, color='blue', edgecolor='black', label='Actual Marks')
plt.plot(x, y, color='red', linewidth=2, label='Normal Distribution Fit')
plt.title('Normal Distribution Curve Fit for Marks', fontsize=14)
plt.xlabel('Marks', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
# Multiple Histograms
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)
plt.hist(data1,alpha=0.5,label='Gorup A',orientation='horizontal')
plt.hist(data2,alpha=0.5,label='Gorup B',orientation='horizontal')
plt.title("Multiple Histogram Representation")
plt.xlabel("Frequency")
plt.ylabel("Data")
plt.legend()
plt.grid()
plt.show()

plt.hist(data1,alpha=0.5,label='Gorup A',orientation='vertical',cumulative=True)
plt.hist(data2,alpha=0.5,label='Gorup B',orientation='vertical',cumulative=True)
plt.title("Multiple Histogram Representation")
plt.xlabel("Frequency")
plt.ylabel("Data")
plt.legend()
plt.grid()
plt.show()
```

```python
data = np.random.normal(50, 10, 200)
plt.figure(figsize=(10,5))
box = plt.boxplot(
    data,
    patch_artist=True,
    showmeans=True
)
for patch in box['boxes']:
    patch.set_facecolor('skyblue')
plt.title("Student Marks Distribution")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

```python
# Sub Plots
x = [1,2,3,4]
y1 = [2,4,6,8]
y2 = [1,3,5,7]
plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Plot 1")
plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Plot 2")
plt.show()
```

```python
data = np.random.randn(100)

fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].plot(data)
ax[0,0].set_title("Line Plot")

ax[0,1].hist(data)
ax[0,1].set_title("Histogram")

ax[1,0].scatter(range(100), data)
ax[1,0].set_title("Scatter Plot")

ax[1,1].boxplot(data)
ax[1,1].set_title("Box Plot")

plt.tight_layout()

plt.show()
```

```python
data = np.random.randn(100)

fig, ax = plt.subplots(2,2, figsize=(10,8))
plt.style.use('dark_background')
fig.suptitle("Sales Dashboard")
ax[0,0].plot(data)
ax[0,0].set_title("Line Plot")

ax[0,1].hist(data)
ax[0,1].set_title("Histogram")

ax[1,0].scatter(range(100), data)
ax[1,0].set_title("Scatter Plot")

ax[1,1].boxplot(data)
ax[1,1].set_title("Box Plot")

plt.tight_layout()

plt.show()
```

```python
plt.style.use('dark_background')

x = np.arange(1,11)
y = [10,15,20,18,25,30,28,35,40,38]

fig, ax = plt.subplots(figsize=(10,5))
# plt.plot(months,sales,color='blue',linestyle='--',linewidth=3,marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='black',alpha=0.8,label='Sales')

ax.plot(
    x,
    y,
    color='blue',
    linewidth=3,
    linestyle='--',
    marker='o',
    markersize=8,
    markerfacecolor='yellow',
    markeredgecolor='black',
    label='Sales'
)

ax.set_title(
    "Monthly Sales Trend",
    fontsize=18,
    fontweight='bold'
)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Revenue", fontsize=12)
ax.grid(
    linestyle='--',
    alpha=0.5
)
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# Contour Plots
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X,Y,Z)
plt.show()
```
