```python
import pandas as pd

new = {
    "Name": ["Amit", "Rahul", "Priya", "Neha"],
    "Marks": [85, 92, 78, 88],
    "City": ["Surat", "Delhi", "Mumbai", "Surat"]
}

df = pd.DataFrame(new)

print(f"Dataset : \n\n{df}\n\n")
print(f"Average Marks : {df['Marks'].mean()}")
print(f"Max Value : {df['Marks'].max()}")
print(f"Min Value : {df['Marks'].min()}\n\n")
print(f"Students Scoring More Than 80 : \n\n{df[df["Marks"] > 80]}\n\n")
print(f"Average Marks City Wise : \n\n{df.groupby("City")["Marks"].mean()}\n\n")
```
