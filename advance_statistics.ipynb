Advance Statistics

```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.weightstats import ztest
from scipy.stats import chi2_contingency
from sklearn.linear_model import LinearRegression
```

```python
marks = np.array([10,20,20,30,40,50,60,70,80])
print(f"Marks               : {marks}")
print(f"Mean Value          : {np.mean(marks)}")
print(f"Median Value        : {np.median(marks)}")
print(f"Mode Value          : {statistics.mode(marks)}") 
print(f"Standard Deviation  : {np.std(marks)}")
print(f"Variance Value      : {np.var(marks)}")
```

```python
# Variance
salary = [25000, 26000, 27000, 28000, 29000]
print(f"Salary Mean : {np.mean(salary)}")
print(f"Salary Variance : {np.var(salary)}")
```

```python
# Quartiles - Quartiles divide data into 4 equal parts.
new = [10,20,30,40,50,60,70]
q1 = np.percentile(new,25)
q2 = np.percentile(new,50)
q3 = np.percentile(new,75)
print(f"Q1 : {q1}")
print(f"Q2 : {q2}")
print(f"Q3 : {q3}")
```

```python
# Skewness
income = pd.Series([20000, 22000, 25000, 27000, 30000, 500000])
print(f"Skewness Of Income : {income.skew()}")
```

```python
# Kurtosis- Kurtosis measures peakness(sharpness) of data.
k = pd.Series([1, 2, 1.5, 2.1, 20])
print(f"Kurtosis : {k.kurt()}")
```

```python
# Bayes Theorem
a = 0.6
pb = 0.7
b = 0.5
print(f"Bayes Theorem : {(pb*a)/b}")
```

```python
# Expected Value - Average expected result of random events.
products = [1, 2, 3]
probabilities = [0.2, 0.5, 0.3]
expected = sum(p * pr for p, pr in zip(products, probabilities))
print(f"Expected Products Purchased : {expected}")
```

Distributions

```python
# Normal Distributions - normal(mean, stddev, total_random_num)
nr = np.random.normal(50,10,1000)
plt.hist(nr)
plt.show()
```

```python
# Binomial Distribution - Used when event has only two outcomes.
# Binomial Distribution data, 10 trials, Probability of success = 0.5, Generate 5 random results
result = np.random.binomial(10, 0.5, 5)
print(result)
```

```python
# Poisson Distribution - Used for counting events occurring in fixed time.
# Poisson Distribution data, 10 random numbers, average value is around 5
result = np.random.poisson(5, 10)
print(result)
```

```python
# Uniform Distribution - All values have equal probability. - np.random.uniform(start, end, total_values)
result = np.random.uniform(1, 10, 5)
print(result)
```

```python
group1 = [45, 50, 55, 60]
group2 = [65, 70, 75, 80]
whole = stats.ttest_ind(group1, group2)
print(whole)
```

```python
data = [70, 72, 75, 78, 80, 82, 85, 90]
z_score, p_value = ztest(data, value=75)
print("Z-Score:", z_score)
print("P-Value:", p_value)
```

```python
# Chi-Square Test - It checks whether two categories are related or independent.
chi = np.array([[30, 20],[10, 40]])
chi2, p, dof, expected = chi2_contingency(chi)
print(f"Chi-Square Value     : {chi2}")
print(f"P Value              : {p}")
print(f"Degrees of Freedom   : {dof}")
print(f"Expected Frequencies : {expected}")
```

```python
# Pearson Correlation - Measures relationship between two variables.
c = pd.DataFrame({'Study': [1, 2, 3, 4, 5],'Marks': [35, 40, 50, 60, 70]})
print(c.corr())
```

```python
# Linear Regression - Predicts one variable using another variable.
x = np.array([[1000], [1200], [1500], [1800], [2000]])
y = np.array([20, 25, 30, 35, 40])
model = LinearRegression()
model.fit(x, y)                            # Train model using data
prediction = model.predict([[2200]])       # Predict price for 2200 sq.ft house
print("Predicted Price:", prediction)
```
